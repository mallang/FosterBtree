```python
import subprocess
import re
import matplotlib.pyplot as plt
import csv
import os

# Define the binary name
binary_name = 'phm_ps_on_off'

# Define the number of runs for each test
num_runs = 5

# Define buffer pool sizes and thread counts to test
buffer_pool_sizes = [1000, 2000, 5000, 10000]
thread_counts = [1, 2, 3, 4, 5, 6, 7, 8]

# Function to build the binary
def build_binary(binary_name):
    build_command = f'cargo build --release --bin {binary_name}'
    process = subprocess.Popen(build_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()

    if stderr:
        print(f"Build Error: {stderr.decode()}")
        return False
    
    print(stdout.decode())
    return True

# Function to run the binary and extract the I/O counts
def run_binary(buffer_pool_size, thread_count, num_runs):
    insert_times_no_ps = []
    insert_times_ps = []
    get_times_no_ps = []
    get_times_ps = []

    for _ in range(num_runs):
        command = f'../target/release/{binary_name} -u -b {buffer_pool_size} -t {thread_count}'
        process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        stdout, stderr = process.communicate()

        if stderr:
            print(f"Error: {stderr.decode()}")
            return None, None, None, None

        stdout = stdout.decode()
        
        for line in stdout.split('\n'):
            if 'Duration Insert without Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    insert_time_no_ps = float(match.group(1))
                    if match.group(2) == 's':
                        insert_time_no_ps *= 1000  # convert s to ms
                    insert_times_no_ps.append(insert_time_no_ps)
            if 'Duration Insert with Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    insert_time_ps = float(match.group(1))
                    if match.group(2) == 's':
                        insert_time_ps *= 1000  # convert s to ms
                    insert_times_ps.append(insert_time_ps)
            if 'Duration Get without Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    get_time_no_ps = float(match.group(1))
                    if match.group(2) == 's':
                        get_time_no_ps *= 1000  # convert s to ms
                    get_times_no_ps.append(get_time_no_ps)
            if 'Duration Get with Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    get_time_ps = float(match.group(1))
                    if match.group(2) == 's':
                        get_time_ps *= 1000  # convert s to ms
                    get_times_ps.append(get_time_ps)

    avg_insert_time_no_ps = sum(insert_times_no_ps) / len(insert_times_no_ps)
    avg_insert_time_ps = sum(insert_times_ps) / len(insert_times_ps)
    avg_get_time_no_ps = sum(get_times_no_ps) / len(get_times_no_ps)
    avg_get_time_ps = sum(get_times_ps) / len(get_times_ps)

    return avg_insert_time_no_ps, avg_insert_time_ps, avg_get_time_no_ps, avg_get_time_ps

# Build the binary
if not build_binary(binary_name):
    print("Failed to build the binary. Exiting.")
    exit(1)

# Automatically assign colors for each buffer pool size
color_cycle = plt.colormaps['tab10']
colors = {size: color_cycle(i / len(buffer_pool_sizes)) for i, size in enumerate(buffer_pool_sizes)}

# CSV file to store results
csv_file = f'{binary_name}.csv'

# Read existing results from the CSV file
existing_results = {}
if os.path.exists(csv_file):
    with open(csv_file, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row
        for row in reader:
            buffer_pool_size, thread_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = map(float, row)
            if buffer_pool_size not in existing_results:
                existing_results[buffer_pool_size] = {}
            existing_results[buffer_pool_size][thread_count] = (insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps)

# Initialize lists to store the results
results = []

# Run the binary for each buffer pool size and thread count, and store the results
for size in buffer_pool_sizes:
    for t_count in thread_counts:
        if size in existing_results and t_count in existing_results[size]:
            # Use existing results
            insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = existing_results[size][t_count]
            results.append([size, t_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps])
        else:
            # Run the command and get new results
            insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = run_binary(size, t_count, num_runs)
            if insert_time_no_ps is not None and insert_time_ps is not None and get_time_no_ps is not None and get_time_ps is not None:
                results.append([size, t_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps])
                print(f"Buffer Pool Size: {size}, Thread Count: {t_count}, Insert Time without PS: {insert_time_no_ps}ms, Insert Time with PS: {insert_time_ps}ms, Get Time without PS: {get_time_no_ps}ms, Get Time with PS: {get_time_ps}ms")

# Write the results to a CSV file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Buffer Pool Size', 'Thread Count', 'Insert Time without PS', 'Insert Time with PS', 'Get Time without PS', 'Get Time with PS'])
    writer.writerows(results)

# Extract data for plotting
insert_times_no_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
insert_times_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
get_times_no_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
get_times_ps = {bp_size: [] for bp_size in buffer_pool_sizes}

for result in results:
    buffer_pool_size, thread_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = result
    if buffer_pool_size in buffer_pool_sizes:
        insert_times_no_ps[buffer_pool_size].append((thread_count, insert_time_no_ps))
        insert_times_ps[buffer_pool_size].append((thread_count, insert_time_ps))
        get_times_no_ps[buffer_pool_size].append((thread_count, get_time_no_ps))
        get_times_ps[buffer_pool_size].append((thread_count, get_time_ps))

# Plot the results for Insert Times
plt.figure(figsize=(12, 6))
for bp_size in buffer_pool_sizes:
    color = colors[bp_size]
    if insert_times_no_ps[bp_size]:
        thread_count_no_ps, insert_time_no_ps = zip(*sorted(insert_times_no_ps[bp_size]))
        plt.plot(thread_count_no_ps, insert_time_no_ps, label=f'Insert without PS (BP Size: {bp_size})', linestyle='--', marker='x', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_no_ps[-1], insert_time_no_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')
    if insert_times_ps[bp_size]:
        thread_count_ps, insert_time_ps = zip(*sorted(insert_times_ps[bp_size]))
        plt.plot(thread_count_ps, insert_time_ps, label=f'Insert with PS (BP Size: {bp_size})', linestyle='-', marker='o', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_ps[-1], insert_time_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')

plt.xlabel('Thread Count')
plt.ylabel('Time (ms)')
plt.title('Insert Times vs. Thread Count for Different Buffer Pool Sizes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot the results for Get Times
plt.figure(figsize=(12, 6))
for bp_size in buffer_pool_sizes:
    color = colors[bp_size]
    if get_times_no_ps[bp_size]:
        thread_count_no_ps, get_time_no_ps = zip(*sorted(get_times_no_ps[bp_size]))
        plt.plot(thread_count_no_ps, get_time_no_ps, label=f'Get without PS (BP Size: {bp_size})', linestyle='--', marker='x', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_no_ps[-1], get_time_no_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')
    if get_times_ps[bp_size]:
        thread_count_ps, get_time_ps = zip(*sorted(get_times_ps[bp_size]))
        plt.plot(thread_count_ps, get_time_ps, label=f'Get with PS (BP Size: {bp_size})', linestyle='-', marker='o', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_ps[-1], get_time_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')

plt.xlabel('Thread Count')
plt.ylabel('Time (ms)')
plt.title('Get Times vs. Thread Count for Different Buffer Pool Sizes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import subprocess
import re
import matplotlib.pyplot as plt
import csv
import os

## Define the binary name
# binary_name = 'phm_ps_on_off'
binary_name = 'phm_ps_skewed_onoff'
## Define whether to build the binary or not
# build_binary_flag = True
build_binary_flag = False
## Define the number of runs for each test
num_runs = 1
# num_runs = 5
# num_runs = 10
## Define buffer pool sizes and thread counts to test
# buffer_pool_sizes = [1000, 2000, 5000, 10000]
# buffer_pool_sizes = [1000, 2000]
buffer_pool_sizes = [5000, 10000]
thread_counts = [1, 2, 3, 4, 5, 6, 7, 8]

# Function to build the binary
def build_binary(binary_name):
    build_command = f'cargo build --release --bin {binary_name}'
    process = subprocess.Popen(build_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()

    if stderr:
        print(f"Build Error: {stderr.decode()}")
        return False
    
    print(stdout.decode())
    return True

# Function to run the binary and extract the I/O counts
def run_binary(buffer_pool_size, thread_count, num_runs):
    insert_times_no_ps = []
    insert_times_ps = []
    get_times_no_ps = []
    get_times_ps = []

    for _ in range(num_runs):
        command = f'../target/release/{binary_name} -u -b {buffer_pool_size} -t {thread_count}'
        process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        stdout, stderr = process.communicate()

        if stderr:
            print(f"Error: {stderr.decode()}")
            return None, None, None, None

        stdout = stdout.decode()
        
        for line in stdout.split('\n'):
            if 'Duration Insert without Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    insert_time_no_ps = float(match.group(1))
                    if match.group(2) == 's':
                        insert_time_no_ps *= 1000  # convert s to ms
                    insert_times_no_ps.append(insert_time_no_ps)
            if 'Duration Insert with Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    insert_time_ps = float(match.group(1))
                    if match.group(2) == 's':
                        insert_time_ps *= 1000  # convert s to ms
                    insert_times_ps.append(insert_time_ps)
            if 'Duration Get without Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    get_time_no_ps = float(match.group(1))
                    if match.group(2) == 's':
                        get_time_no_ps *= 1000  # convert s to ms
                    get_times_no_ps.append(get_time_no_ps)
            if 'Duration Get with Pointer Swizzling:' in line:
                match = re.search(r'([\d.]+)(ms|s)', line)
                if match:
                    get_time_ps = float(match.group(1))
                    if match.group(2) == 's':
                        get_time_ps *= 1000  # convert s to ms
                    get_times_ps.append(get_time_ps)

    avg_insert_time_no_ps = sum(insert_times_no_ps) / len(insert_times_no_ps)
    avg_insert_time_ps = sum(insert_times_ps) / len(insert_times_ps)
    avg_get_time_no_ps = sum(get_times_no_ps) / len(get_times_no_ps)
    avg_get_time_ps = sum(get_times_ps) / len(get_times_ps)

    return avg_insert_time_no_ps, avg_insert_time_ps, avg_get_time_no_ps, avg_get_time_ps

# Build the binary if the flag is set
if build_binary_flag:
    if not build_binary(binary_name):
        print("Failed to build the binary. Exiting.")
        exit(1)

# Automatically assign colors for each buffer pool size
color_cycle = plt.colormaps['tab10']
colors = {size: color_cycle(i / len(buffer_pool_sizes)) for i, size in enumerate(buffer_pool_sizes)}

# Define line styles for swizzling on/off
line_styles = {
    'no_ps': '--',
    'ps': '-'
}

# CSV file to store results
csv_file = f'{binary_name}.csv'

# Read existing results from the CSV file
existing_results = {}
all_results = []
if os.path.exists(csv_file):
    with open(csv_file, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row
        for row in reader:
            buffer_pool_size, thread_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = map(float, row)
            all_results.append([buffer_pool_size, thread_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps])
            if buffer_pool_size not in existing_results:
                existing_results[buffer_pool_size] = {}
            existing_results[buffer_pool_size][thread_count] = (insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps)

# Run the binary for each buffer pool size and thread count, and store the results
for size in buffer_pool_sizes:
    for t_count in thread_counts:
        if size in existing_results and t_count in existing_results[size]:
            # Use existing results
            insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = existing_results[size][t_count]
            all_results.append([size, t_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps])
        else:
            # Run the command and get new results
            insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = run_binary(size, t_count, num_runs)
            if insert_time_no_ps is not None and insert_time_ps is not None and get_time_no_ps is not None and get_time_ps is not None:
                all_results.append([size, t_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps])
                print(f"Buffer Pool Size: {size}, Thread Count: {t_count}, Insert Time without PS: {insert_time_no_ps}ms, Insert Time with PS: {insert_time_ps}ms, Get Time without PS: {get_time_no_ps}ms, Get Time with PS: {get_time_ps}ms")

# Write the results to a CSV file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Buffer Pool Size', 'Thread Count', 'Insert Time without PS', 'Insert Time with PS', 'Get Time without PS', 'Get Time with PS'])
    writer.writerows(all_results)

# Extract data for plotting
insert_times_no_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
insert_times_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
get_times_no_ps = {bp_size: [] for bp_size in buffer_pool_sizes}
get_times_ps = {bp_size: [] for bp_size in buffer_pool_sizes}

for result in all_results:
    buffer_pool_size, thread_count, insert_time_no_ps, insert_time_ps, get_time_no_ps, get_time_ps = result
    if buffer_pool_size in buffer_pool_sizes:
        insert_times_no_ps[buffer_pool_size].append((thread_count, insert_time_no_ps))
        insert_times_ps[buffer_pool_size].append((thread_count, insert_time_ps))
        get_times_no_ps[buffer_pool_size].append((thread_count, get_time_no_ps))
        get_times_ps[buffer_pool_size].append((thread_count, get_time_ps))

# Plot the results for Insert Times
plt.figure(figsize=(12, 6))
for bp_size in buffer_pool_sizes:
    color = colors[bp_size]
    if insert_times_no_ps[bp_size]:
        thread_count_no_ps, insert_time_no_ps = zip(*sorted(insert_times_no_ps[bp_size]))
        plt.plot(thread_count_no_ps, insert_time_no_ps, label=f'Insert without PS (BP Size: {bp_size})', linestyle=line_styles['no_ps'], marker='x', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_no_ps[-1], insert_time_no_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')
    if insert_times_ps[bp_size]:
        thread_count_ps, insert_time_ps = zip(*sorted(insert_times_ps[bp_size]))
        plt.plot(thread_count_ps, insert_time_ps, label=f'Insert with PS (BP Size: {bp_size})', linestyle=line_styles['ps'], marker='o', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_ps[-1], insert_time_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')

plt.xlabel('Thread Count')
plt.ylabel('Time (ms)')
plt.title('Insert Times vs. Thread Count for Different Buffer Pool Sizes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot the results for Get Times
plt.figure(figsize=(12, 6))
for bp_size in buffer_pool_sizes:
    color = colors[bp_size]
    if get_times_no_ps[bp_size]:
        thread_count_no_ps, get_time_no_ps = zip(*sorted(get_times_no_ps[bp_size]))
        plt.plot(thread_count_no_ps, get_time_no_ps, label=f'Get without PS (BP Size: {bp_size})', linestyle=line_styles['no_ps'], marker='x', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_no_ps[-1], get_time_no_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')
    if get_times_ps[bp_size]:
        thread_count_ps, get_time_ps = zip(*sorted(get_times_ps[bp_size]))
        plt.plot(thread_count_ps, get_time_ps, label=f'Get with PS (BP Size: {bp_size})', linestyle=line_styles['ps'], marker='o', color=color)
        # Add annotation for the line
        plt.annotate(f'BP Size: {bp_size}', (thread_count_ps[-1], get_time_ps[-1]), textcoords="offset points", xytext=(10,-10), ha='center')

plt.xlabel('Thread Count')
plt.ylabel('Time (ms)')
plt.title('Get Times vs. Thread Count for Different Buffer Pool Sizes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
